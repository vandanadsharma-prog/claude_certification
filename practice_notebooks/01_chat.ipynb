```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6" 
```

```python
chat_messages = []

def append_message(role, message):
    new_message = {"role": role, "content": message}
    chat_messages.append(new_message)
    return chat_messages
```

```python
def chat(message: str, system_prompt: str, temperature : float, stop_sequences : list =[]):
    messages = append_message("user", message)
    response = client.messages.create(model=model, max_tokens=1024, messages=messages, system=system_prompt, temperature=temperature, stop_sequences=stop_sequences)
    reply = response.content[0].text
    print(reply)
    append_message("assistant", reply)
    return reply
```

```python
def chat_stream(message: str, system_prompt: str, temperature : float):
    messages = append_message("user", message)
    with client.messages.stream(model=model, 
                                max_tokens=1024, 
                                messages=messages, 
                                system=system_prompt, 
                                temperature=temperature) as stream:
        for text in stream.text_stream:
            print(text)
            append_message("assistant", text)
            return text
```

```python
temperature = 0.1   # --> dial this upto 100 for creativity
system_prompt = "You are a data engineer with expertise on data science"
while True:
    user_message = input(">")
    if user_message.lower() == "exit":
        break
    # chat(user_message, system_prompt, temperature)
    text += chat_stream(user_message, system_prompt, temperature)
    print("----------------------------------------------------")
```
