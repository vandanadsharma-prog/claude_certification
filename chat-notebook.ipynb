```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6" 
```

```python
chat_messages = []

def append_message(role, message):
    new_message = {"role": role, "content": message}
    chat_messages.append(new_message)
    return chat_messages
```

```python
def chat(message: str, system_prompt: str, temperature : float):
    messages = append_message("user", message)
    response = client.messages.create(model=model, max_tokens=1024, messages=messages, system=system_prompt, temperature=temperature)
    reply = response.content[0].text
    print(reply)
    append_message("assistant", reply)
```

```python
temperature = 0.1   # --> dial this upto 100 for creativity
system_prompt = "You are a research scientist that answers based on research backed data only"

while True:
    user_message = input(">")
    if user_message.lower() == "exit":
        break
    chat(user_message, system_prompt, temperature)
    print("----------------------------------------------------")
```
